```python
channels = ['ch1', 'ch2', 'ch3', 'ch4', 'ch5', 'ch6', 'ch7', 'ch8', 'ch9', 'ch10', 'ch11', 'ch12', 'ch13', 'ch14', 'ch15', 'ch16']
epochs = ["epoch" + str(i+1) for i in range(17280)]
```

```python
df = pd.DataFrame.from_records(reshaped_data).T
df.columns = channels
df
```

```python
df['epochs'] = epochs
df['packet_loss_idx'] = packet_loss_idx
```

```python
df
```

```python
d = []
for epoch in df.epochs: # df.epochs
    e = df[df.epochs == epoch]
    #print(epoch)
    a = []
    for channel in channels:
        arr = e[channel].to_numpy()
        a.append(arr)
    d.append(a)
    
    
df_d = pd.DataFrame(d, columns=channels) 
df_d['combined'] = df_d.values.tolist()
df_d
```

```python
df_combined = df_d[['combined']]
df_combined
```

```python
for i in df_combined.T:
    df_combined.at[i, 'combined'] = functools.reduce(operator.iconcat, df_combined.at[i, 'combined'], [])  
df_combined
```

#### The time domain features will be calculated for each epoch for each channel (and the channels determined by the correlation matrices).  
Time domain features: mean value, median, standard deviation, root mean square (RMS), kurtosis, skewness, variance, max, min etc.

#### MEAN

```python
df_mean = df_combined.applymap(lambda x: np.mean(x)).assign(epochs=epochs).rename(columns={"combined": "mean"})
df_mean
```

#### MEDIAN

```python
df_median = df_combined.applymap(lambda x: np.median(x)).assign(epochs=epochs).rename(columns={"combined": "median"})
df_median
```

#### VARIANCE

```python
df_variance = df_combined.applymap(lambda x: np.var(x)).assign(epochs=epochs).rename(columns={"combined": "variance"})
df_variance
```

#### SKEWNESS

```python
df_skewness = df_combined.applymap(lambda x: skew(x)).assign(epochs=epochs).rename(columns={"combined": "skewness"})
df_skewness
```

#### KURTOSIS

```python
df_kurtosis = df_combined.applymap(lambda x: kurtosis(x)).assign(epochs=epochs).rename(columns={"combined": "kurtosis"})
df_kurtosis
```

#### MIN

```python
df_min = df_combined.applymap(lambda x: np.min(x)).assign(epochs=epochs).rename(columns={"combined": "min"})
df_min
```

#### MAX

```python
df_max = df_combined.applymap(lambda x: np.min(x)).assign(epochs=epochs).rename(columns={"combined": "max"})
df_max
```

#### STANDARD DEVIATION

```python
df_stdev = df_combined.applymap(lambda x: np.std(x)).assign(epochs=epochs).rename(columns={"combined": "stdev"})
df_stdev
```

```python
X = pd.merge(df_mean, df_median, on='epochs').merge(pd.merge(df_variance, df_stdev, on='epochs')).merge(pd.merge(df_skewness, df_kurtosis, on='epochs')).merge(pd.merge(df_min, df_max, on='epochs'))
```
