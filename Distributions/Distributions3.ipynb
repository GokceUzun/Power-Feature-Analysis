### Stratified Group K-Folds

```python
# STRATIFIED GROUP 5-FOLD CROSS VALIDATION SPLITS

sgkf1 = StratifiedGroupKFold(n_splits=5)

for i, (train_index, test_index) in enumerate(sgkf1.split(X1, y1, groups1)):
    print(f"Fold {i}:")
    print(f"TRAIN: index={train_index}")
    print(f"train_group={groups1[train_index]}")
    print("\n")
    print(f"VAL:  index={test_index}")
    print(f"val_group={groups1[test_index]}")
    print("\n")
    print("\n")
```

```python
# CLASS DISTRIBUTION
# 5-fold stratified group cv
# Three classes 0 wake and 1 non-rem

distribution = pd.DataFrame(columns=['0', '1'])
indx = []
for i, (train_index, test_index) in enumerate(sgkf1.split(X1, y1, groups1)):
    X_train, y_train =  X1.iloc[train_index.tolist()], y1.iloc[train_index.tolist()]
    X_val, y_val = X1.iloc[test_index.tolist()], y1.iloc[test_index.tolist()]
    #print(y_train.value_counts(normalize=True))
    #print(y_val.value_counts(normalize=True))
    #print("\n")
    distribution = pd.concat([ distribution, y_train.value_counts(normalize=True).to_frame().T], axis=0)
    indx.append(f'training set - fold {i}')
    distribution = pd.concat([ distribution, y_val.value_counts(normalize=True).to_frame().T], axis=0)
    indx.append(f'validation set - fold {i}')
    
distribution.index = indx
distribution.iloc[:,2:]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats 
import numpy as np  
from scipy.stats import norm
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

### Two classes

```python
for i, col in enumerate(X1.columns.to_list()):
    plt.figure(figsize=(8,4))
    #plt.subplot(12, 8, i + 1)
    data = X1[col]
    mu, std = norm.fit(data)
    plt.hist(data, bins=15, density=True, rwidth=100, alpha=0.6, color='b')

    # Plot the PDF.
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2)
    title = col + " (mu = %.2f std = %.2f)" % (mu, std)
    plt.title(title)

    plt.show()
```

```python
# Normalize data between 0 and 1

scaler = MinMaxScaler()
norm_X1 = scaler.fit_transform(X1)

for i, col in enumerate(X1.columns.to_list()):
    plt.figure(figsize=(8,4))
    #plt.subplot(12, 8, i + 1)
    data = norm_X1[i]
    mu, std = norm.fit(data)
    plt.hist(data, bins=15, density=True, alpha=0.6, color='b')

    # Plot the PDF.
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2)
    title = col + " (mu = %.2f std = %.2f)" % (mu, std)
    plt.title(title)

    plt.show()
```
